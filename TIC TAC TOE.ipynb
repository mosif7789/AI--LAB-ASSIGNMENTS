Using MiniMax Algorithm

```python
import math

board = [
    ['_', '_', '_'],
    ['_', '_', '_'],
    ['_', '_', '_']
]

def evaluate(b):
    for row in b:
        if row == ['O', 'O', 'O']:
            return 10
        if row == ['X', 'X', 'X']:
            return -10
    return 0


def minimax(board, depth, isMax):
    score = evaluate(board)

    if score == 10 or score == -10:
        return score

    if isMax:
        best = -math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'O'
                    best = max(best, minimax(board, depth + 1, False))
                    board[i][j] = '_'

        return best

    else:
        best = math.inf

        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'X'
                    best = min(best, minimax(board, depth + 1, True))
                    board[i][j] = '_'

        return best
```

using Alpha Beta Pruning

```python
import math

def alphabeta(depth, nodeIndex, maximizingPlayer, values, alpha, beta):

    # If we reached leaf node
    if depth == 3:
        return values[nodeIndex]

    if maximizingPlayer:
        best = -math.inf

        for i in range(2):
            val = alphabeta(depth + 1, nodeIndex * 2 + i, False, values, alpha, beta)
            best = max(best, val)
            alpha = max(alpha, best)

            if beta <= alpha:
                break

        return best

    else:
        best = math.inf

        for i in range(2):
            val = alphabeta(depth + 1, nodeIndex * 2 + i, True, values, alpha, beta)
            best = min(best, val)
            beta = min(beta, best)

            if beta <= alpha:
                break

        return best


values = [3, 5, 9, 1, 2, 0, -1]

result = alphabeta(0, 0, True, values, -math.inf, math.inf)

print("Optimal value:", result)
```
